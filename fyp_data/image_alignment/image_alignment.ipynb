```python
def get_blob(path):
    image = Image.open(path)
    buffer = io.BytesIO()
    image.save(buffer, format="PNG")
    return buffer.getvalue()
```

```python
image_mapping_list = []
```

```python
image_mapping_list = []  # Ensure this list is initialized outside all loops

for i in range(1, 111):
    chinese_directory = Path(f"../squares/{i}/chinese")
    sumerian_directory = Path(f"../squares/{i}/sumerian")

    chinese_image_files = list(chinese_directory.iterdir())
    sumerian_image_files = list(sumerian_directory.iterdir())

    for chinese_image_file in chinese_image_files:
        chinese_image = get_blob(chinese_image_file)

        for sumerian_image_file in sumerian_image_files:
            sumerian_image = get_blob(sumerian_image_file)

            # Generate a fresh UUID and new dictionary
            image_mapping_dict = {
                "_id": str(uuid.uuid4()),
                "number": i,
                "chinese_image": chinese_image,
                "sumerian_image": sumerian_image,
            }

            # Append to the list
            image_mapping_list.append(image_mapping_dict)
```
