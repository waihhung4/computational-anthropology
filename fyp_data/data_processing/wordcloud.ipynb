```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

chinese_keyword_list = ["old china", "ancient china", "shang dynasty", "xia dynasty"]
sumerian_keyword_list = ["sumerian", "sumer", "sumerian", "sumerian civilization", "sumerian history"]
combined_keyword_list = ["ancient culture", "china and sumer", "sumerian and chinese", "china sumer", "sumerian chinese"]

chinese_stopwords = STOPWORDS.union({'china', 'chinese','ancient', 'civilization', 'culture', 'people', 'i', 'history', 'mean', 'old', 'use', 'way', 'use'})
sumerian_stopwords = STOPWORDS.union({'ancient', 'sumerian','sumer', "sumerians" 'civilization',  'culture', 'people', 'i', 'history', 'mean', 'old', 'use', 'way'})
combined_stopwords = {'china', 'chinese', 'ancient', 'sumerian','sumer', "sumerians", 'civilization', 'culture', 'people', 'i', 'history', 'mean', 'old', 'use', 'way'}
```

<h1>ALL<h1>

```python
preprocessed_content_data = get_all_content()
df = pd.DataFrame(preprocessed_content_data)
preprocessed_text_list = df["preprocessed_token"].to_list()
flat_tokens = [' '.join(doc) for doc in preprocessed_text_list]
text = ' '.join(flat_tokens).lower()

wordcloud = WordCloud(width=800, height=400, background_color='white', 
                       stopwords=combined_stopwords, max_words=100, colormap='coolwarm').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
