```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

chinese_keyword_list = ["old china", "ancient china", "shang dynasty", "xia dynasty"]
sumerian_keyword_list = ["sumerian", "sumer", "sumerian", "sumerian civilization", "sumerian history"]
combined_keyword_list = ["ancient culture", "china and sumer", "sumerian and chinese", "china sumer", "sumerian chinese"]

chinese_stopwords = STOPWORDS.union({'china', 'chinese','ancient', 'civilization', 'culture', 'people'})
sumerian_stopwords = STOPWORDS.union({'ancient', 'sumerian','sumer', "sumerians" 'civilization',  'culture', 'people'})
combined_stopwords = STOPWORDS.union({'china', 'chinese', 'ancient', 'sumerian','sumer', "sumerians", 'civilization', 'culture', 'people'})
```

<h1>ALL<h1>

```python
preprocessed_content_data = get_all_preprocessed_trial()
df = pd.DataFrame(preprocessed_content_data)
preprocessed_text_list = df["preprocessed_token"].to_list()
flat_tokens = [' '.join(doc) for doc in preprocessed_text_list]  # Join each noun phrase in a document
text = ' '.join(flat_tokens)  # Combine all the documents into one large string

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', 
                       stopwords=combined_keyword_list, max_words=100, colormap='coolwarm').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
